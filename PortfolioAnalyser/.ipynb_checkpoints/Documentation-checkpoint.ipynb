````sos
import pandas as pd
import matplotlib as plt
import matplotlib.style
from pandas_datareader import data as pdr

from matplotlib.ticker import FuncFormatter

import yfinance as yf
yf.pdr_override()

pd.set_option("display.max_rows", 20)

# Set matplotlib style
plt.style.use('seaborn')

def merge_time_series(df_1, df_2):
    df = df_1.merge(df_2, how='left', left_index=True, right_index=True)
    return df

def merge_time_series_outer(df_1, df_2):
    df = df_1.merge(df_2, how='outer')
    return df
````

```python
    # Make a copy of the dataframe
    df_shifted = df.copy()

```

````sos
def rolling_years_return(df, years):
    
    # Make a copy of the dataframe
    df_shifted = df.copy()    
    # Shift by one year
    df_shifted = df.shift(freq=pd.DateOffset(years=1))    
    # Reindex shifted dataframe
    date_time = pd.DataFrame(index = df.index)
    df_shifted = merge_time_series(date_time, df_shifted)
    
    # Return division of original dataframe by shifted dataframe
    df =  (((df / df_shifted) - 1)).loc[df_shifted.index[0]:]
    return df
````

### Download SPY quotes

````sos
%%capture
# download quotes
tickers = ['SPY']
Quotes = pd.DataFrame()
Start ='2005-08-01'
End = "2020-01-15"

for ticker in tickers:
    Quotes[ticker] = pdr.get_data_yahoo(ticker, start=Start, end=End)['Adj Close']
````

````sos
rolling_one_year_return = rolling_years_return(Quotes, years=1)
rolling_one_year_return.tail(10)
````

````sos
ax = rolling_one_year_return.plot(figsize=(15, 8), lw=1.5, color = ['royalblue'], alpha =0.9, title='1 year SPY rolling return')
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y))) 
````

````sos
### Comparing to 252 session return
````

````sos
def multi_period_return(df, years = 1, days=252):
    shifted = df.shift(days * years)
    One_year = (((1 + (df - shifted) / shifted) ** (1 / years))-1)
    return One_year
````

````sos
MPR = multi_period_return(Quotes).dropna()
MPR
````

````sos
df = merge_time_series(rolling_one_year_return, MPR)
df.columns=['ROYR', 'MPR']
df
````

````sos
ax = df.dropna().plot(figsize=(15, 8), lw=1.5, color = ['royalblue', 'grey'], alpha =0.9, title='1 year SPY rolling return vs 252 session return')
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y))) 
````
